```python
N = 5
men_means = (6.854788, 6.391415, 3.188267, 10.514267, 5.877909)
#men_std = (2, 3, 4, 1, 2)

ind = np.arange(N)  # the x locations for the groups
width = 0.25    # the width of the bars

fig, ax = plt.subplots(figsize=(12,8))
rects1 = ax.bar(ind, men_means, width, color='wheat')

women_means = (8.319637, 2.538806, 3.311113, 10.422983, 5.636023)
#women_std = (3, 5, 2, 3, 3)
rects2 = ax.bar(ind + width, women_means, width, color='tomato')

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('Scores by Wine')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('fixed acidity', 'residual sugar', 'pH', 'alcohol', 'quality'))

ax.legend((rects1[0], rects2[0]), ('White', 'Red'))

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,'%d' % int(height), ha='center')

autolabel(rects1)
autolabel(rects2)
```

## Remove Outliers for Prediction
Using White Wine data

```python
# For each feature find the data points with extreme high or low values
for feature in white.keys():

    # TODO: Calculate Q1 (25th percentile of the data) for the given feature
    Q1 = np.percentile(white[feature], q=25)
 
    # TODO: Calculate Q3 (75th percentile of the data) for the given feature
    Q3 = np.percentile(white[feature], q=75)
 
    # TODO: Use the interquartile range to calculate an outlier step (1.5 times the interquartile range)
    interquartile_range = Q3 - Q1
    step = 1.5 * interquartile_range
 
    # Display the outliers
    print("Data points considered outliers for the feature '{}':".format(feature))
    display(white[~((white[feature] >= Q1 - step) & (white[feature] <= Q3 + step))])
```
